# K-Means

K-Means algorithm is a centroid based clustering algorithm where the sum of distances of points from the centroid of each cluster is minimized. The final output is a set of K clusters .ie. the cluster assigned to each point and the K centroids of the clusters.

### The Algorithm
> 1.   Initialize K centroids to K points randomly and set each point's initial cluster as the centroid closest to it.
2.   While the clusers are changing - \\
a. Assign the new centroids as the centroids of the points which have the same assigned cluster. \\
b. Assign the new clusters to the points as the closest centroid.
3. Return the assignments and the centroids.





### Implementation of KMeans and testing on Synthetic Datasets

```python
# Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Extending the color list to accommodate more clusters
color=["darkturquoise","darkorange","teal","darkviolet","tomato", "yellowgreen","hotpink","gold", "brown", "gray"]
```

```python
# Generating Synthetic Data with different parameters

k=3
color=["darkturquoise","darkorange","teal","darkviolet","tomato", "yellowgreen","hotpink","gold"]

X_new, _ = make_blobs(n_samples=700, n_features=3, centers=k, random_state=20)
print("Shape = " + str(X_new.shape))

print("Sample:")
print(X_new[:5])
```

```python
#DEFINE ALL THE FUNCTIONS & VARIABLES WE NEED TO DO CLUSTERING
#Create an empty dictionary to store the cluster name, centroid and points associated with the cluster.
clusters={}

def initializeClusterCentroids(X, k, color):
  #For each cluster, initialize its properties
  for i in range(k):
      center = X[np.random.choice(X.shape[0])] # Initialize centroids with random data points
      points = []
      cluster = {  'center' : center,
          'points' : points,
          'color' : color[i]
      }
      clusters[i] = cluster #looping over clusters dictionary

#Calculate euclidean distance between v1 and v2
def distance(v1, v2):
    return np.sum((v2-v1)**2)**0.5

def assignPointsToCluster(clusters, X):
    for ix in range(X.shape[0]):#iterate for all datapoints
        #Foe each datapoint, find the distance to k centroids
        distance_of_i = []
        for kx in range(len(clusters)):
            d = distance(X[ix], clusters[kx]['center']) #we have find distance b/w
            distance_of_i.append(d)
        #here we want find the minimum distance of cluster centers
        cluster_to_choose = np.argmin(distance_of_i)
        clusters[cluster_to_choose]['points'].append(X[ix])

#STEP 3: UPDATE CLUSTER CENTROIDS
def updateCluster(clusters):
    for kx in range(len(clusters)):
        cluster_points = clusters[kx]['points']
        cluster_points = np.array(cluster_points)
        #here pts are array of list but we want array of array so
        if len(cluster_points)>0:
            new_center = np.mean(cluster_points, axis=0) #axis is along rows so will
            #find mean of all the feature seprately it will give 2,0 when we have cluster pts 50,2
            clusters[kx]['center'] = new_center
            clusters[kx]['points'] = [] # clear the points in a cluster list (emptying pts)
            #as we have to do step 2 after step 3 again till not converge
```

```python
#We need to visualize this too. Lets define a function that can do that!
def plotclusters(clusters, plot):
    for kx in range(k):
        cluster_points = clusters[kx]['points']
        cluster_color = clusters[kx]['color']
        cluster_center = clusters[kx]['center']
        cluster_points = np.array(cluster_points)

        #plotting points associated /nearest to the cluster centers
        if len(cluster_points) > 0:
          plot.scatter(cluster_points[:, 0], cluster_points[:, 1], s = 2, c = cluster_color)

        plot.scatter(cluster_center[0], cluster_center[1], s = 250, c = cluster_color, marker="o")
```

We can now run and visualize how the clusters evolve.

```python
fig, axs = plt.subplots(2, 3)
fig.set_size_inches(15, 10)
axs[0, 0].scatter(X_new[:,0],X_new[:,1], s = 2, c = 'black')
axs[0, 0].set_title('Dataset')

np.random.seed(1)
clusters = {} # Clear the clusters dictionary before initialization
initializeClusterCentroids(X_new, k, color)

#Now, lets plot the centroids it has initalized
axs[0, 1].set_title('Centroid Initialization')
axs[0, 1].scatter(X_new[:, 0], X_new[:, 1], s = 2, c = 'black')
for i in range(k):#for all the clusters
    center = clusters[i]['center'] #its cluster's center
    axs[0, 1].scatter(center[0], center[1], c=clusters[i]['color'], s=250, marker="o")

#CLUSTERING EPOCH 1
axs[0, 2].set_title('Iteration 1')
assignPointsToCluster(clusters, X_new)
plotclusters(clusters, axs[0, 2])
updateCluster(clusters)

#CLUSTERING EPOCH 2
axs[1, 0].set_title('Iteration 2')
assignPointsToCluster(clusters, X_new)
plotclusters(clusters, axs[1, 0])
updateCluster(clusters)

#CLUSTERING EPOCH 3
axs[1, 1].set_title('Iteration 3')
assignPointsToCluster(clusters, X_new)
plotclusters(clusters, axs[1, 1])
updateCluster(clusters)

#CLUSTERING EPOCH 4
axs[1, 2].set_title('Iteration 4')
assignPointsToCluster(clusters, X_new)
plotclusters(clusters, axs[1, 2])
updateCluster(clusters)

plt.tight_layout()
plt.show()
```

Hence our KMeans algo can be run by calling the following function with appropriate X (dataset)

```python
from copy import deepcopy

def run_kmeans(k, X, color):
  clusters = {} # Clear the clusters dictionary before initialization for each run
  initializeClusterCentroids(X, k, color)
  assignPointsToCluster(clusters, X)
  while (True):
    prev_clusters = deepcopy(clusters)
    updateCluster(clusters)
    assignPointsToCluster(clusters, X)
    # Check for convergence by comparing the centers of the clusters
    converged = True
    for i in range(k):
        if not np.array_equal(prev_clusters[i]['center'], clusters[i]['center']):
            converged = False
            break
    if converged:
      break
  return clusters
```

### KMeans on Digits Dataset (KMeans for labelling)

Here we will attempt to use k-means to try to identify similar digits without using the original label information; this might be similar to a first step in extracting meaning from a new dataset about which you don't have any a priori label information.

```python
from sklearn.datasets import load_digits

digits = load_digits()
print(digits.data.shape)
```

```python
from sklearn.cluster import KMeans

# Perform KMeans
kmeans = KMeans(n_clusters=10, random_state=0, n_init=10)
clusters = kmeans.fit_predict(digits.data)
print(kmeans.cluster_centers_.shape)
```

```python
# We can now see how the centers of the clusters formed are

fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

Because k-means knows nothing about the identity of the cluster, the 09 labels may be permuted. We can fix this by matching each learned cluster label with the true labels found in them:

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

Now we can check how accurate our unsupervised clustering was in finding similar digits within the data:

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```

**Hence clustering algorithms like KMeans are highly effective for identifying labels for groups apriori and even a simple KMeans acheives a ~80% accuracy on digit classification.**

The confusion matrix for this looks like:

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()

mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

### KMeans on Iris Dataset (Elbow Method)

We will perform KMeans on the famous Iris Dataset which has four features: sepal length, sepal width, petal length, and petal width. The fifth column is for species, which holds the value for these types of plants.

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from sklearn import datasets

#import the dataset
iris = datasets.load_iris()
# print(iris.data[:10])
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
x = df.iloc[:, [0,1,2,3]].values
display(df.head(5))
```

Let us start with assigning the initial number of clusters as k=5.

```python
kmeans5 = KMeans(n_clusters=5, n_init=10) # Added n_init for clarity
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)

print(kmeans5.cluster_centers_) # Added print to display centers
```

How do we choose the optimal number of clusters? Theres a method called the Elbow method, which is designed to help find the optimal number of clusters in a dataset. So lets use this method to calculate the optimum value of k. To implement the Elbow method, we need to create some Python code (shown below), and well plot a graph between the number of clusters and the corresponding error value.

This graph generally ends up shaped like an elbow, hence its name:

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, n_init=10) # Added n_init for clarity
    kmeans.fit(x)
    Error.append(kmeans.inertia_)

import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

The output graph of the Elbow method shows that the shape of elbow is approximately formed at k=3. The optimal value of k is between 2 and 4, as the elbow-like shape is formed at k=3 in the above graph.

```python
kmeans3 = KMeans(n_clusters=3, n_init=10)
y_kmeans3 = kmeans3.fit_predict(x)
print(y_kmeans3)

print(kmeans3.cluster_centers_)
```

We can now visualize the clustering in two dimensions instead of the original four.

```python
plt.scatter(x[:,0], x[:,1], c=y_kmeans3, cmap='rainbow')
```

1.  **Is feature scaling essential for KMeans as it is for most ML algos? Explain.**

    Yes, feature scaling is generally essential for KMeans. KMeans is a distance-based algorithm. It calculates the distance between data points and cluster centroids. If the features have different scales, features with larger values will have a disproportionately larger impact on the distance calculation compared to features with smaller values. This can lead to skewed results where the clustering is heavily influenced by features with wider ranges, regardless of their actual importance. Scaling ensures that all features contribute equally to the distance calculation.

2.  **What are ways to prevent initialization variation in KMeans?**

    The initialization of centroids in KMeans can significantly impact the final clustering result. Here are a few ways to prevent initialization variation:

    *   **Run KMeans multiple times with different random initializations:** The most common approach is to run the KMeans algorithm several times with different random centroid starting points and choose the best result based on a metric like inertia (within-cluster sum of squares). The `n_init` parameter in scikit-learn's `KMeans` does exactly this.
    *   **Use a smart initialization method like K-Means++:** K-Means++ is an initialization technique that selects initial centroids in a way that is likely to lead to a better final clustering. It selects the first centroid randomly, and then selects subsequent centroids with a probability proportional to their squared distance from the closest existing centroid. This tends to spread out the initial centroids. Scikit-learn's `KMeans` uses 'k-means++' by default.
    *   **Initialize centroids based on domain knowledge:** If you have prior knowledge about the data and potential clusters, you might be able to manually select initial centroid locations.

3.  **What is the training and testing complexity of KMeans?**

    The complexity of the KMeans algorithm is typically discussed in terms of training complexity, as it is an unsupervised learning algorithm and doesn't have a separate testing phase in the same way supervised learning algorithms do.

    The training complexity of KMeans is roughly O(k * n * d * i), where:

    *   k is the number of clusters.
    *   n is the number of data points.
    *   d is the number of features (dimensions).
    *   i is the number of iterations required for convergence.

    In practice, KMeans often converges relatively quickly, so 'i' is often much smaller than 'n'. The complexity is linear with respect to the number of data points, features, and clusters. However, the number of iterations 'i' can vary depending on the data and initialization.

    For testing (assigning new data points to existing clusters after training), the complexity is O(k * d) for each new data point, as you just need to calculate the distance to each of the k centroids. For assigning a set of m new data points, the complexity is O(m * k * d).

## Agglomerative Hierarchical Clustering

In this technique, initially each data point is considered as an individual cluster. At each iteration, the similar clusters merge with other clusters until one cluster or K clusters are formed.

The basic algorithm of Agglomerative is straight forward.

    Compute the proximity matrix
    Let each data point be a cluster
    Repeat: Merge the two closest clusters and update the proximity matrix
    Until only a single cluster remains

To understand better lets see a pictorial representation of the Agglomerative Hierarchical clustering Technique. Lets say we have six data points {A,B,C,D,E,F}.

    Step- 1: In the initial step, we calculate the proximity of individual points and consider all the six data points as individual clusters as shown in the image below.

image.png

    Step- 2: In step two, similar clusters are merged together and formed as a single cluster. Lets consider B,C, and D,E are similar clusters that are merged in step two. Now, were left with four clusters which are A, BC, DE, F.
    Step- 3: We again calculate the proximity of new clusters and merge the similar clusters to form new clusters A, BC, DEF.
    Step- 4: Calculate the proximity of the new clusters. The clusters DEF and BC are similar and merged together to form a new cluster. Were now left with two clusters A, BCDEF.
    Step- 5: Finally, all the clusters are merged together and form a single cluster.

You cut the dendrogram tree with a horizontal line at a height where the line can traverse the maximum distance up and down without intersecting the merging point.

## Divisive Hierarchical Clustering

In Divisive or DIANA(DIvisive ANAlysis Clustering) is a top-down clustering method where we assign all of the observations to a single cluster and then partition the cluster to two least similar clusters. Finally, we proceed recursively on each cluster until there is one cluster for each observation. So this clustering approach is exactly opposite to Agglomerative clustering.

image.png

There is evidence that divisive algorithms produce more accurate hierarchies than agglomerative algorithms in some circumstances but is conceptually more complex.

In both agglomerative and divisive hierarchical clustering, users need to specify the desired number of clusters as a termination condition(when to stop merging).

Divisive clustering methods are fairly complicated to implement, so we will not do that here. You can check out open source implementations like https://github.com/div338/Divisive-Clustering-Analysis-Program-DIANA- for further details.

1.  **What is the need for hierarchical clustering and its advantages over KMeans?**

    Hierarchical clustering is needed when you want to understand the hierarchical relationships between your data points and visualize these relationships. Unlike KMeans, which requires you to specify the number of clusters beforehand, hierarchical clustering builds a tree-like structure (dendrogram) that shows how data points merge or split at different levels of similarity.

    **Advantages of Hierarchical Clustering over KMeans:**

    *   **No need to pre-specify the number of clusters (k):** You can decide on the number of clusters by cutting the dendrogram at a desired level.
    *   **Provides a dendrogram:** The dendrogram visualizes the clustering process and the relationships between clusters at different levels, offering more insight into the data structure.
    *   **Can reveal nested clusters:** Hierarchical clustering can identify clusters that are nested within larger clusters.
    *   **Less sensitive to the initial placement of centroids:** Unlike KMeans, which can be sensitive to initial centroid placement, hierarchical clustering's outcome is deterministic once a distance metric and linkage method are chosen.

    However, hierarchical clustering can be computationally more expensive than KMeans, especially for large datasets.

2.  **What are the advantages of Density Based Clustering over KMeans?**

    Density-based clustering algorithms, like DBSCAN, are advantageous when you have clusters of arbitrary shapes and sizes, and when your data contains noise or outliers. KMeans assumes that clusters are spherical and of roughly equal size, and it is sensitive to outliers.

    **Advantages of Density Based Clustering over KMeans:**

    *   **Can discover clusters of arbitrary shapes and sizes:** Density-based methods can find clusters that are not spherical, unlike KMeans.
    *   **Can identify noise and outliers:** These algorithms can classify points that do not belong to any cluster as noise.
    *   **Does not require specifying the number of clusters beforehand:** Similar to hierarchical clustering, the number of clusters is determined by the algorithm based on the density parameters.
    *   **Less sensitive to outliers:** Outliers are typically identified as noise rather than forcing them into existing clusters, which can happen with KMeans.

    Density-based clustering requires careful selection of density parameters (e.g., epsilon and minimum points), which can sometimes be challenging.
